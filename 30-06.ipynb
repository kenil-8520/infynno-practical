```python
while True:
    print("Tic Tac Toe Game")
    print("1. Play")
    print("2. View Score")
    print("0. Exit")

    choice1 = int(input("enter number "))
    
    if choice1 == 1:
        l = [' ',' ',' ',' ',' ',' ',' ',' ', ' ',' ']
        game = 0
        user = 1
        tie = 0
        win = 1

        def draw():
            
            print(" %c | %c | %c " % (l[1],l[2],l[3]))
            print("___|___|___")
            print(" %c | %c | %c " % (l[4],l[5],l[6]))
            print("___|___|___")
            print(" %c | %c | %c " % (l[7],l[8],l[9]))
            print("   |   |   ")
        # draw()

        def check(x):
            if l[x] == ' ':
                return True
            else:
                return False

        def winner():
            global game
            if l[1] == l[2] and l[2] == l[3] and l[1] !=' ':
                game = win
            elif l[1] == l[5] and l[5] == l[9] and l[1] !=' ':
                game = win
            elif l[1] == l[4] and l[4] == l[7] and l[1] !=' ':
                game = win
            elif l[4] == l[5] and l[5] == l[6] and l[4] !=' ':
                game = win
            elif l[7] == l[8] and l[8] == l[9] and l[7] !=' ':
                game = win
            elif l[2] == l[5] and l[5] == l[8] and l[2] !=' ':
                game = win
            elif l[3] == l[5] and l[5] == l[7] and l[3] !=' ':
                game = win
            elif l[3] == l[6] and l[6] == l[9] and l[3] !=' ':
                game = win
            elif l[1]!=' ' and l[2]!=' ' and l[3]!=' ' and l[4]!=' ' and l[5]!=' ' and l[6]!=' ' and l[7]!=' ' and l[8]!=' ' and l[9]!=' ':
                game=tie
            else:
                game=0

        # draw()
        while(game==0):
            draw()    
            if(user % 2 != 0):
                print("Player X's chance")
                mark = 'X'
            else:
                print("Player O's chance")
                mark = 'O'
            choice = int(input("Enter the position between  1 - 9 : "))

            if(check(choice)):
                l[choice] = mark
                user+=1
                winner()

        draw()
        global c1
        global c2 
        if(game==win):
            user-=1
            if(user%2!=0):
                c1+=1
                print("Player X Won")
            else:
                c2+=1
                print("Player O Won")
        elif(game==tie):
            print("game tie")

    elif choice1 == 2:
        def result():
            print("___________________")
            print("|______Score______|")
            print("|   X    |    O   |")
            print(f"|   {c1}    |    {c2}   |")
            print("|________|________|")
        result()  
    
    elif choice1 == 0:  
        c1 = 0
        c2 = 0
        break
    else:
        print("invalid choice")
```
